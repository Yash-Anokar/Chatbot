```python
from langchain_community.document_loaders import TextLoader
loadder  = TextLoader('Cpp_Introduction.txt',encoding="utf-8")
loadder
```

```python
text_documents = loadder.load()
text_documents
```

```python
# Text splitter
from langchain_text_splitters import RecursiveCharacterTextSplitter
textSplitter = RecursiveCharacterTextSplitter(chunk_size=200,chunk_overlap=20)
finalDocuments = textSplitter.split_documents(text_documents)
finalDocuments
```

```python
# If finalDocuments is a list of Document objects:
texts = [doc.page_content for doc in finalDocuments]
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings  = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
print(embeddings)
```

```python
from langchain_community.vectorstores import FAISS
db = FAISS.from_documents(finalDocuments, embeddings)
db
```

```python
query = input("Enter your Question for C++: ")
docs = db.similarity_search(query)
for i in range(len(docs)):
    print(docs[i].page_content)
```
